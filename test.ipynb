```python
import pandas_datareader as pdweb
```

```python
import datetime
```

```python
prices = pdweb.get_data_yahoo(['CVX', 'XOM', 'BP'],
                             start=datetime.datetime(2010,1,1),
                             end=datetime.datetime(2013,1,1))['Adj Close']
```

```python
prices.head()
```

```python
returns = prices.pct_change() #pacentage_change
```
